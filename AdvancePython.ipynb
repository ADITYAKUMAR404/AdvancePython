```python
# Library Management System

library = []
def add_book():
    book_id = input("Enter Book ID: ")
    title = input("Enter Book Title: ")
    author = input("Enter Author Name: ")

    book = {
        "id": book_id,
        "title": title,
        "author": author,
        "issued": False
    }

    library.append(book)
    print("Book added successfully!\n")

def display_books():
    if len(library) == 0:
        print("Library is empty.\n")
    else:
        print("\nLibrary Books:")
        for book in library:
            status = "Issued" if book["issued"] else "Available"
            print(f'ID: {book["id"]}, Title: {book["title"]}, Author: {book["author"]}, Status: {status}')
        print()

def issue_book():
    book_id = input("Enter Book ID to issue: ")
    for book in library:
        if book["id"] == book_id:
            if not book["issued"]:
                book["issued"] = True
                print("Book issued successfully!\n")
            else:
                print("Book is already issued.\n")
            return
    print("Book not found.\n")

def return_book():
    book_id = input("Enter Book ID to return: ")
    for book in library:
        if book["id"] == book_id:
            if book["issued"]:
                book["issued"] = False
                print("Book returned successfully!\n")
            else:
                print("Book was not issued.\n")
            return
    print("Book not found.\n")

def menu():
    while True:
        print("====== Library Menu ======")
        print("1. Add Book")
        print("2. Display Books")
        print("3. Issue Book")
        print("4. Return Book")
        print("5. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_book()
        elif choice == "2":
            display_books()
        elif choice == "3":
            issue_book()
        elif choice == "4":
            return_book()
        elif choice == "5":
            print("Exiting Library System...")
            break
        else:
            print("Invalid choice. Try again.\n")


# Run the program
menu()
```

```python
import time

def traffic_light():
    try:
        while True:
            # RED Signal
            print("\nRED Light - STOP")
            for i in range(3, 0, -1):
                print("Time Remaining:", i, "seconds", end="\r")
                time.sleep(1)

            # GREEN Signal
            print("\nGREEN Light - GO")
            for i in range(3, 0, -1):
                print("Time Remaining:", i, "seconds", end="\r")
                time.sleep(1)

            # YELLOW Signal
            print("\nYELLOW Light - WAIT")
            for i in range(3, 0, -1):
                print("Time Remaining:", i, "seconds", end="\r")
                time.sleep(1)

    except KeyboardInterrupt:
        print("\n\nTraffic Light System Ended.")

# Run the system
traffic_light()
```

```python
# Parking Managment System
class ParkingLot:
    def __init__(self, capacity):
        self.capacity = capacity
        self.spaces = {}  # {slot_number: (vehicle_number, vehicle_type)}
        self.vehicles = {}  # {vehicle_number: vehicle_type}

    def park_vehicle(self, v_no, v_type):
        if v_no in self.vehicles:
            print(f"Vehicle {v_no} is already parked!")
            return

        # Find empty slot
        for slot in range(1, self.capacity + 1):
            if slot not in self.spaces:
                self.spaces[slot] = (v_no, v_type)
                self.vehicles[v_no] = v_type
                print(f"Vehicle {v_no} ({v_type}) parked at slot {slot}")
                return

        print("Parking Lot is Full!")

    def remove_vehicle(self, v_no):
        if v_no not in self.vehicles:
            print(f"Vehicle {v_no} not found in parking lot!")
            return

        # Find and remove the vehicle
        for slot, (vehicle_no, vehicle_type) in self.spaces.items():
            if vehicle_no == v_no:
                del self.spaces[slot]
                del self.vehicles[v_no]
                print(f"Vehicle {v_no} removed from slot {slot}")
                return

    def show_status(self):
        if not self.spaces:
            print("Parking lot is empty")
            return

        print("\nParking Lot Status:")
        print("-" * 40)
        for slot in sorted(self.spaces.keys()):
            v_no, v_type = self.spaces[slot]
            print(f"Slot {slot}: {v_no} ({v_type})")
        print(f"Available Slots: {self.capacity - len(self.spaces)}")

    def show_statistics(self):
        print("\nParking Lot Statistics:")
        print("-" * 40)
        print(f"Total Capacity: {self.capacity}")
        print(f"Occupied Slots: {len(self.spaces)}")
        print(f"Available Slots: {self.capacity - len(self.spaces)}")
        print(f"Occupancy Rate: {(len(self.spaces) / self.capacity) * 100:.1f}%")


# Create Parking Lot with 10 spaces
p1 = ParkingLot(10)

while True:
    print("\n1. Park Vehicle")
    print("2. Remove Vehicle")
    print("3. Show Status")
    print("4. Show Statistics")
    print("5. Exit")

    choice = input("Enter choice: ")

    if choice == "1":
        v_no = input("Enter Vehicle Number: ")
        v_type = input("Enter Vehicle Type (Car/Bike): ")
        p1.park_vehicle(v_no, v_type)

    elif choice == "2":
        v_no = input("Enter Vehicle Number to Remove: ")
        p1.remove_vehicle(v_no)

    elif choice == "3":
        p1.show_status()

    elif choice == "4":
        p1.show_statistics()

    elif choice == "5":
        print("Exiting System...")
        break

    else:
        print("Invalid Choice!")

```

```python
#TO_DO LIST
tasks = []
def add_task():
    title = input("Enter task title: ")
    category = input("Enter category/project: ")
    deadline = input("Enter deadline (YYYY-MM-DD): ")
    priority = input("Enter priority (High/Medium/Low): ")

    task = {
        "title": title,
        "category": category,
        "deadline": deadline,
        "priority": priority,
        "status": "Pending"
    }

    tasks.append(task)
    print("Task added successfully!\n")

def view_tasks():
    if not tasks:
        print("No tasks available.\n")
        return

    for i, task in enumerate(tasks):
        print(f"{i+1}. {task['title']} | Category: {task['category']} | Deadline: {task['deadline']} | Priority: {task['priority']} | Status: {task['status']}")
    print()

def complete_task():
    view_tasks()
    num = int(input("Enter task number to mark completed: ")) - 1

    if 0 <= num < len(tasks):
        tasks[num]["status"] = "Completed"
        print("Task marked as completed!\n")
    else:
        print("Invalid task number.\n")

def filter_category():
    cat = input("Enter category to filter: ")

    found = False
    for task in tasks:
        if task["category"].lower() == cat.lower():
            print(f"{task['title']} | Deadline: {task['deadline']} | Priority: {task['priority']} | Status: {task['status']}")
            found = True

    if not found:
        print("No tasks found for this category.\n")
    print()
def menu():
    while True:
        print("===== TO-DO LIST MANAGER =====")
        print("1. Add Task")
        print("2. View Tasks")
        print("3. Complete Task")
        print("4. Filter by Category")
        print("5. Exit")

        choice = input("Enter choice: ")

        if choice == "1":
            add_task()
        elif choice == "2":
            view_tasks()
        elif choice == "3":
            complete_task()
        elif choice == "4":
            filter_category()
        elif choice == "5":
            print("Exiting program...")
            break
        else:
            print("Invalid choice\n")


menu()
```

```python
# Warehouse Management System

warehouse = {}

# Add Product
def add_product():
    pid = input("Enter Product ID: ")
    name = input("Enter Product Name: ")
    quantity = int(input("Enter Quantity: "))
    price = float(input("Enter Price: "))

    warehouse[pid] = {
        "name": name,
        "quantity": quantity,
        "price": price
    }

    print("Product added successfully!\n")


# View Products
def view_products():
    if not warehouse:
        print("Warehouse is empty\n")
    else:
        print("\n--- Product List ---")
        for pid, details in warehouse.items():
            print("Product ID:", pid)
            print("Name:", details["name"])
            print("Quantity:", details["quantity"])
            print("Price:", details["price"])
            print("----------------------")


# Update Stock
def update_stock():
    pid = input("Enter Product ID to update: ")

    if pid in warehouse:
        quantity = int(input("Enter new quantity: "))
        warehouse[pid]["quantity"] = quantity
        print("Stock updated successfully\n")
    else:
        print("Product not found\n")


# Delete Product
def delete_product():
    pid = input("Enter Product ID to delete: ")

    if pid in warehouse:
        del warehouse[pid]
        print("Product deleted successfully\n")
    else:
        print("Product not found\n")


# Search Product
def search_product():
    pid = input("Enter Product ID to search: ")

    if pid in warehouse:
        details = warehouse[pid]
        print("\nProduct Found")
        print("Name:", details["name"])
        print("Quantity:", details["quantity"])
        print("Price:", details["RS.price"])
    else:
        print("Product not found\n")


# Main Menu
def menu():
    while True:
        print("\n===== Warehouse Management System =====")
        print("1. Add Product")
        print("2. View Products")
        print("3. Update Stock")
        print("4. Delete Product")
        print("5. Search Product")
        print("6. Exit")

        choice = input("Enter your choice: ")

        if choice == "1":
            add_product()
        elif choice == "2":
            view_products()
        elif choice == "3":
            update_stock()
        elif choice == "4":
            delete_product()
        elif choice == "5":
            search_product()
        elif choice == "6":
            print("Exiting program...")
            break
        else:
            print("Invalid choice")


# Run program
menu()
```
